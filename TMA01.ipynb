```python
from matplotlib import pyplot as plt
import numpy as np
from scipy.interpolate import make_interp_spline
```

```python
x = np.array(range(0, 620, 40))
for n in x: #prints all values of x so that I can check they are correct
    print (n)
```

```python
y = np.array([0, 6, 12, 20, 28, 36, 40, 40, 40, 40, 38, 30, 20, 10, 4, 0])
```

```python
#plot points
plt.plot (x, y, 'bs')

#define x as 200 equally spaced values between the min and max of original x
xnew = np.linspace(x.min(), x.max(), 200)

#define spline
spl = make_interp_spline(x, y, k=3)
y_smooth = spl(xnew)

#create smooth line chart
plt.plot(xnew, y_smooth)

#create labels
plt.xlabel ("t /s")
plt.ylabel ("$v_x$" + " " + "$/ms^{-1}$")
plt.title ("A graph showing the velocity over time of a train from Oxford\nto Didcot, based on data contained in the TMA question.")

#save as svg for TMA file
plt.savefig("tma01_q2_graph.png")

#show plot
plt.show()
```
