```python
from matplotlib import pyplot as plt
import numpy as np
```

First, add the data from the TMA question to the notebook. Time is the independent variable, so will be plotted on the x-axis.

```python
x = np.array(range(0, 620, 40))
for n in x: #prints all values of x so that I can check they are correct
    print (n)
```

```python
y = np.array([0, 6, 12, 20, 28, 36, 40, 40, 40, 40, 38, 30, 20, 10, 4, 0])
```

```python
plt.plot (x, y, 'bs')
lines = plt.plot (x, y)
plt.setp (lines, color='b', linewidth =1.0)
plt.xlabel ("t /s")
plt.ylabel ("$v_x$" + " " + "$/ms^{-1}$")
plt.title ("A graph showing the velocity over time of a train from Oxford to Didcot, based on data contained in the TMA question.")
plt.show()
```
